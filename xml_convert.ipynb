```python
import pyogrio
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
```

```python
buildings = gpd.read_file("Building_Footprint.kml", driver="KML")
```

```python
buildings = buildings.to_crs(epsg=4326)
```

```python
buildings
```

```python
df = pd.read_csv('Training_data_uhi_index.csv')
```

```python
df
```

```python
geometry = [Point(xy) for xy in zip(df.Longitude, df.Latitude)]
points = gpd.GeoDataFrame(df, geometry=geometry, crs="EPSG:4326")
```

```python
points
```

```python
joined = gpd.sjoin(points, buildings, how="left", predicate="within")
```

```python
joined
```

```python
joined['has_building'] = joined.index_right.notnull().astype(int)
```

```python
joined
```
